```python
import pandas as pd

movies = pd.read_csv(r"D:\ml_25m_project\ml_25m_project\movies_preprocess.csv")
```

```python
movies.info()
```

```python
movies.describe()
```

```python
#按 movieId 分组，计算每部电影的平均评分
average_rating = movies.groupby(['movieId'])['rating'].mean()
#获取用户平均评分最高的前10部电影
top10_average_rating = average_rating.sort_values(ascending=False).head(10)
print(top10_average_rating)
```

```python
rating_5_count = (average_rating == 5.0).sum()
print(f"平均评分为5.0的电影数量为{rating_5_count}部")
rating_05_count = (average_rating == 0.5).sum()
print(f"平均评分为0.5的电影数量为{rating_05_count}部")
```

```python
#统计每个用户推荐的电影总数
user_recommend_movie_count = movies.groupby('userId')['movieId'].count()
#按电影数量降序排序并取推荐电影数量最多的 10 位用户
top10_users_recommend_movies = user_recommend_movie_count.sort_values(ascending=False).head(10)
print("推荐电影数量最多的 10 位用户ID和具体数量")
print(top10_users_recommend_movies)
#按电影数量降序排序并取推荐数量最少的 10 位用户
bottom10_users_recommend_movies = user_recommend_movie_count.sort_values(ascending=True).head(10)
print("推荐电影数量最少的 10 位用户ID和具体数量")
print(bottom10_users_recommend_movies)
```

```python
#统计仅推荐 20 部电影的用户数量
recommenddation_20_count = (user_recommend_movie_count == 20).sum()
print(f"仅推荐 20 部电影的用户数量为{recommenddation_20_count}位")
```

```python
#去重 movieId，避免重复电影
unique_movies = movies.drop_duplicates(subset=['movieId'])
#统计每年电影的数量并获取电影数量最多的前10年
movies_num_of_year = unique_movies.groupby('year')['movieId'].count()
top10_movies_num_of_year = movies_num_of_year.sort_values(ascending=False).head(10)
print(top10_movies_num_of_year)
```
